# Tutorial 1:  Basic usage of the adaptive Local Knockoff Filter

This tutorial demonstrates a basic usage example of the adaptive Local Knockoff Filter (aLKF) to test for local effects with binary covariates.

```r
set.seed(2022)
```

## Load a toy example data set

```r
suppressMessages(library(tidyverse))
source("../LKF/aLKF.R")

# Load the simulated data set
ifile <- "data/data_n2000_p100_a400-0_i2_delta5_linear_glmnet_b1.txt"
data <- read_delim(ifile, delim=" ", col_types=cols())
X <- select(data, starts_with("X_")) %>% data.matrix()
Y <- select(data, starts_with("Y")) %>% data.matrix() %>% as.numeric()
Z <- select(data, starts_with("Z_")) %>% data.matrix()

# Load the knockoff variables
ifile <- "data/knockoffs_n2000_p100_a400-0_i2_delta5_linear_glmnet_b1.txt"
Xk <- read_delim(ifile, delim=" ", col_types=cols()) %>% select(starts_with("Xk")) %>% data.matrix()

# Load the ground-truth individual effects
ifile <- "data/ite_n2000_p100_a400-0_i2_delta5_linear_glmnet_b1.txt"
ite.data <- read_delim(ifile, delim=" ", col_types=cols()) %>% data.matrix()
ite <- ite.data

# Print basic information about the data
cat(sprintf("This data set contains %d observations of %d binary explanatory variables and %d covariates.\n", length(Y), ncol(X), ncol(Z)))
```

```r
hist(Y)
```

## Apply the adaptive Local Knockoff Filter

To apply the adaptive local knockoff filter to test for local effects, we begin by specifying which covariates are allowed to interact with the explanatory variables. In this example, we consider only interactions between the explanatory variables and the binary covariates.

```r
# The first 60 covariates are continuous
Z.noint <- Z[,1:60]

# The remaining 20 covariates are discrete
Z.int <- Z[,61:ncol(Z)]
```

Below, we apply the aLKF at the nominal 10\% FDR level to detect and test up to 2 different interactions per variable.

```r
res.aLKF <- aLKF_analysis(Y, X, Xk, Z.noint, Z.int, ite=ite, family = "gaussian", fdr.nominal=0.1, fdr.offset=1,
                        naive=FALSE, split=FALSE, KFglobal=FALSE, cross.prior=TRUE, num.int=2, verbose=FALSE)
```

## Evaluate the discoveries

Using our knowledge of the ground-truth individual effects, we can measure the performance of the aLKF in terms of power, false discovery proportion (FDP), and homogeneity of the tested hypotheses.

```r
discoveries.aLKF <- res.aLKF$discoveries
hypotheses.aLKF <- res.aLKF$hypotheses
evaluate_results(discoveries.aLKF, hypotheses.aLKF)
```

The full list of discoveries reported by the aLKF is:
